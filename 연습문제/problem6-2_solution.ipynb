```python
from typing import TypedDict, Literal, List
from langgraph.graph import StateGraph, START, END, MessagesState
from langchain_core.messages import HumanMessage, AIMessage
from langchain_community.vectorstores import FAISS
from langchain_ollama import OllamaEmbeddings
from langchain_core.documents import Document
import random
import re

# 1ë‹¨ê³„: ìƒíƒœ ì •ì˜ (MessagesState í™•ì¥)
class CafeInquiryState(MessagesState):
    inquiry_type: str           # ë¬¸ì˜ ìœ í˜•
    retrieved_menus: List[str]  # ê²€ìƒ‰ëœ ë©”ë‰´ë“¤

# Vector DB ì´ˆê¸°í™”
embeddings_model = OllamaEmbeddings(model="qwen2.5:1.5b")
menu_db = FAISS.load_local(
    "./db/cafe_db", 
    embeddings_model, 
    allow_dangerous_deserialization=True
)

def extract_menu_info(doc: Document) -> dict:
    """ë¬¸ì„œì—ì„œ ë©”ë‰´ ì •ë³´ ì¶”ì¶œ"""
    content = doc.page_content
    menu_name = doc.metadata.get('menu_name', 'Unknown')
    
    # ê°€ê²© ì¶”ì¶œ
    price_match = re.search(r'â‚©([\d,]+)', content)
    price = price_match.group(0) if price_match else "ê°€ê²© ì •ë³´ ì—†ìŒ"
    
    # ì„¤ëª… ì¶”ì¶œ
    description_match = re.search(r'ì„¤ëª…:\s*(.+?)(?:\n|$)', content, re.DOTALL)
    description = description_match.group(1).strip() if description_match else "ì„¤ëª… ì—†ìŒ"
    
    return {
        "name": menu_name,
        "price": price,
        "description": description,
        "full_content": content
    }

# 2ë‹¨ê³„: ë…¸ë“œ í•¨ìˆ˜ë“¤ ì •ì˜

def classify_inquiry(state: CafeInquiryState) -> CafeInquiryState:
    """ì‚¬ìš©ì ë¬¸ì˜ë¥¼ ë¶„ë¥˜í•˜ëŠ” ë‹¨ê³„"""
    print("---ë¬¸ì˜ ìœ í˜• ë¶„ë¥˜---")
    
    last_message = state["messages"][-1].content
    
    # í‚¤ì›Œë“œ ê¸°ë°˜ ë¶„ë¥˜
    if any(keyword in last_message for keyword in ["ê°€ê²©", "ì–¼ë§ˆ", "ë¹„ìš©"]):
        inquiry_type = "ê°€ê²© ë¬¸ì˜"
    elif any(keyword in last_message for keyword in ["ì¶”ì²œ", "ë­ê°€ ì¢‹", "ì–´ë–¤ ê²Œ"]):
        inquiry_type = "ì¶”ì²œ ìš”ì²­"
    else:
        inquiry_type = "ë©”ë‰´ ë¬¸ì˜"
    
    print(f"ë¶„ë¥˜ëœ ë¬¸ì˜ ìœ í˜•: {inquiry_type}")
    
    return {"inquiry_type": inquiry_type}

def handle_price_inquiry(state: CafeInquiryState) -> CafeInquiryState:
    """ê°€ê²© ë¬¸ì˜ì— ëŒ€í•œ ì‘ë‹µ ìƒì„± (Vector DB ê²€ìƒ‰)"""
    print("---ê°€ê²© ë¬¸ì˜ ì²˜ë¦¬ (Vector DB ê²€ìƒ‰)---")
    
    # ì „ì²´ ë©”ë‰´ ê²€ìƒ‰ì„ ìœ„í•œ ì¼ë°˜ì ì¸ ì¿¼ë¦¬
    docs = menu_db.similarity_search("ë©”ë‰´ ê°€ê²©", k=5)
    
    response = "ì €í¬ ì¹´í˜ì˜ ë©”ë‰´ë³„ ê°€ê²©ì€ ë‹¤ìŒê³¼ ê°™ìŠµë‹ˆë‹¤:\n\n"
    
    if docs:
        for doc in docs:
            menu_info = extract_menu_info(doc)
            response += f"â€¢ {menu_info['name']}: {menu_info['price']}\n"
    else:
        response += "ì£„ì†¡í•©ë‹ˆë‹¤. í˜„ì¬ ê°€ê²© ì •ë³´ë¥¼ ë¶ˆëŸ¬ì˜¬ ìˆ˜ ì—†ìŠµë‹ˆë‹¤."
    
    response += "\ní˜¹ì‹œ íŠ¹ì • ë©”ë‰´ì— ëŒ€í•´ ë” ìì„¸í•œ ì •ë³´ê°€ í•„ìš”í•˜ì‹œë©´ ì–¸ì œë“  ë§ì”€í•´ ì£¼ì„¸ìš”!"
    
    # ê²€ìƒ‰ëœ ë©”ë‰´ ì •ë³´ ì €ì¥
    retrieved_menus = [extract_menu_info(doc)["name"] for doc in docs] if docs else []
    
    return {
        "messages": [AIMessage(content=response)],
        "retrieved_menus": retrieved_menus
    }

def handle_recommendation(state: CafeInquiryState) -> CafeInquiryState:
    """ì¶”ì²œ ìš”ì²­ì— ëŒ€í•œ ì‘ë‹µ ìƒì„± (Vector DB ê²€ìƒ‰)"""
    print("---ì¶”ì²œ ìš”ì²­ ì²˜ë¦¬ (Vector DB ê²€ìƒ‰)---")
    
    # ì‚¬ìš©ì ë©”ì‹œì§€ì—ì„œ í‚¤ì›Œë“œ ì¶”ì¶œí•˜ì—¬ ë” ì •í™•í•œ ê²€ìƒ‰
    user_message = state["messages"][-1].content
    
    # ì¶”ì²œ ê´€ë ¨ í‚¤ì›Œë“œë¡œ ê²€ìƒ‰
    docs = menu_db.similarity_search(user_message, k=3)
    
    if not docs:
        # ê¸°ë³¸ ê²€ìƒ‰
        docs = menu_db.similarity_search("ì¸ê¸° ë©”ë‰´", k=3)
    
    response = "ê³ ê°ë‹˜ê»˜ ì¶”ì²œë“œë¦¬ëŠ” ë©”ë‰´ëŠ” ë‹¤ìŒê³¼ ê°™ìŠµë‹ˆë‹¤:\n\n"
    
    if docs:
        # ëœë¤ìœ¼ë¡œ 2ê°œ ì„ íƒ
        selected_docs = random.sample(docs, min(2, len(docs)))
        
        for doc in selected_docs:
            menu_info = extract_menu_info(doc)
            response += f"ğŸ”¸ {menu_info['name']} ({menu_info['price']})\n"
            response += f"   {menu_info['description'][:100]}...\n\n"
        
        response += "ì´ ë©”ë‰´ë“¤ì€ Vector DB ê²€ìƒ‰ì„ í†µí•´ ê³ ê°ë‹˜ì˜ ë¬¸ì˜ì™€ ê°€ì¥ ê´€ë ¨ì„± ë†’ì€ ë©”ë‰´ë¡œ ì„ ì •ë˜ì—ˆìŠµë‹ˆë‹¤!"
    else:
        response += "ì£„ì†¡í•©ë‹ˆë‹¤. í˜„ì¬ ì¶”ì²œí•  ìˆ˜ ìˆëŠ” ë©”ë‰´ ì •ë³´ë¥¼ ë¶ˆëŸ¬ì˜¬ ìˆ˜ ì—†ìŠµë‹ˆë‹¤."
    
    # ê²€ìƒ‰ëœ ë©”ë‰´ ì •ë³´ ì €ì¥
    retrieved_menus = [extract_menu_info(doc)["name"] for doc in docs] if docs else []
    
    return {
        "messages": [AIMessage(content=response)],
        "retrieved_menus": retrieved_menus
    }

def handle_menu_inquiry(state: CafeInquiryState) -> CafeInquiryState:
    """ì¼ë°˜ ë©”ë‰´ ë¬¸ì˜ì— ëŒ€í•œ ì‘ë‹µ ìƒì„± (Vector DB ê²€ìƒ‰)"""
    print("---ë©”ë‰´ ë¬¸ì˜ ì²˜ë¦¬ (Vector DB ê²€ìƒ‰)---")
    
    # ì‚¬ìš©ì ë©”ì‹œì§€ë¥¼ ë°”íƒ•ìœ¼ë¡œ ê²€ìƒ‰
    user_message = state["messages"][-1].content
    docs = menu_db.similarity_search(user_message, k=4)
    
    if not docs:
        # ê¸°ë³¸ ê²€ìƒ‰
        docs = menu_db.similarity_search("ì¹´í˜ ë©”ë‰´", k=4)
    
    response = "ì €í¬ ì¹´í˜ì—ì„œëŠ” ë‹¤ìŒê³¼ ê°™ì€ ë©”ë‰´ë¥¼ ì œê³µí•˜ê³  ìˆìŠµë‹ˆë‹¤:\n\n"
    
    if docs:
        for doc in docs:
            menu_info = extract_menu_info(doc)
            response += f"â€¢ {menu_info['name']}: {menu_info['description'][:80]}... ({menu_info['price']})\n\n"
        
        response += "ìœ„ ë©”ë‰´ë“¤ì€ Vector DB ê²€ìƒ‰ì„ í†µí•´ ê³ ê°ë‹˜ì˜ ë¬¸ì˜ì™€ ê´€ë ¨ì„±ì´ ë†’ì€ ë©”ë‰´ë“¤ì…ë‹ˆë‹¤."
    else:
        response += "ì£„ì†¡í•©ë‹ˆë‹¤. í˜„ì¬ ë©”ë‰´ ì •ë³´ë¥¼ ë¶ˆëŸ¬ì˜¬ ìˆ˜ ì—†ìŠµë‹ˆë‹¤."
    
    response += "\n\nê° ë©”ë‰´ì— ëŒ€í•œ ìì„¸í•œ ì •ë³´ë‚˜ ì¶”ì²œì´ í•„ìš”í•˜ì‹œë©´ ì–¸ì œë“  ë§ì”€í•´ ì£¼ì„¸ìš”!"
    
    # ê²€ìƒ‰ëœ ë©”ë‰´ ì •ë³´ ì €ì¥
    retrieved_menus = [extract_menu_info(doc)["name"] for doc in docs] if docs else []
    
    return {
        "messages": [AIMessage(content=response)],
        "retrieved_menus": retrieved_menus
    }

def decide_response_type(state: CafeInquiryState) -> Literal["price", "recommendation", "menu"]:
    """ë¬¸ì˜ ìœ í˜•ì— ë”°ë¼ ë‹¤ìŒ ì‹¤í–‰í•  ë…¸ë“œë¥¼ ê²°ì •"""
    inquiry_type = state["inquiry_type"]
    if inquiry_type == "ê°€ê²© ë¬¸ì˜":
        return "price"
    elif inquiry_type == "ì¶”ì²œ ìš”ì²­":
        return "recommendation"
    else:
        return "menu"

# 3ë‹¨ê³„: ê·¸ë˜í”„ êµ¬ì„±
builder = StateGraph(CafeInquiryState)

builder.add_node("classify", classify_inquiry)
builder.add_node("handle_price", handle_price_inquiry)
builder.add_node("handle_recommendation", handle_recommendation)
builder.add_node("handle_menu", handle_menu_inquiry)

builder.add_edge(START, "classify")
builder.add_conditional_edges(
    "classify",
    decide_response_type,
    {
        "price": "handle_price",
        "recommendation": "handle_recommendation",
        "menu": "handle_menu"
    }
)
builder.add_edge("handle_price", END)
builder.add_edge("handle_recommendation", END)
builder.add_edge("handle_menu", END)

graph = builder.compile()

# 4ë‹¨ê³„: ì‹¤í–‰ ë° í…ŒìŠ¤íŠ¸
def test_inquiry_system_with_vectordb():
    """Vector DB ì—°ë™ ë¬¸ì˜ ì‹œìŠ¤í…œ í…ŒìŠ¤íŠ¸ í•¨ìˆ˜"""
    test_messages = [
        "ë©”ë‰´ ê°€ê²©ì´ ê¶ê¸ˆí•´ìš”",
        "ì»¤í”¼ ì¤‘ì—ì„œ ì–´ë–¤ ë©”ë‰´ë¥¼ ì¶”ì²œí•´ ì£¼ì‹œë‚˜ìš”?",
        "ë””ì €íŠ¸ ë©”ë‰´ê°€ ìˆë‚˜ìš”?"
    ]

    for message in test_messages:
        print(f"\n{'='*60}")
        print(f"ê³ ê° ë¬¸ì˜: {message}")
        print('='*60)
        
        initial_state = {
            "messages": [HumanMessage(content=message)],
            "inquiry_type": "",
            "retrieved_menus": []
        }
        
        result = graph.invoke(initial_state)
        
        print(f"\nì¹´í˜ ì‘ë‹µ:\n{result['messages'][-1].content}")
        print(f"\nê²€ìƒ‰ëœ ë©”ë‰´ë“¤: {result['retrieved_menus']}")

# í…ŒìŠ¤íŠ¸ ì‹¤í–‰
test_inquiry_system_with_vectordb()
```
